```python
def ip_to_geo(ip):
    with geoip2.database.Reader('c:\\users\\dunai\\appdata\\local\\programs\\python\\python310\\lib\\site-packages\\geoip2\\GeoIP2-City.mmdb') as reader:
        try:
            rcity = reader.city(ip)
            return rcity.city.name, rcity.city
        except geoip2.errors.AddressNotFoundError as e:
            return None

df['Destination City'] = df['Destination IP Address'].apply(ip_to_geo)
```

```python
#Добудем новую информацию по геолокации ip адресов источников и получателей

def ip_to_city(ip):
    db = 'c:\\users\\dunai\\appdata\\local\\programs\\python\\python310\\lib\\site-packages\\geoip2\\SxGeoCity.dat'
    info = sxgeo.SxGeo(db_file=db, mode=sxgeo.SXGEO_BATCH)
    try:
        rcity = info.get_city(ip, full=True)
        if rcity['city']['id'] != '':
            ncity = f"{rcity['city']['name_en']}, {rcity['city']['id']}"
        else:
            ncity = None
        #nregion = rcity['region']['name_en']
        #print(ncity)
        return ncity
    except:
        return None

def ip_to_region(ip):
    db = 'c:\\users\\dunai\\appdata\\local\\programs\\python\\python310\\lib\\site-packages\\geoip2\\SxGeoCity.dat'
    info = sxgeo.SxGeo(db_file=db, mode=sxgeo.SXGEO_BATCH)
    try:
        rcity = info.get_city(ip, full=True)
        if rcity['region']['id'] != '':
            ncity = f"{rcity['region']['name_en']}, {rcity['region']['id']}"
        else:
            ncity = None
        #nregion = rcity['region']['name_en']
        #print(ncity)
        return ncity
    except:
        return None

df['Destination ID City'] = df['Destination IP Address'].apply(ip_to_city)
df['Destination ID Region'] = df['Destination IP Address'].apply(ip_to_region)
df['Source ID City'] = df['Source IP Address'].apply(ip_to_city)
df['Source ID Region'] = df['Source IP Address'].apply(ip_to_region)
```

```python
df = df.dropna()
```

```python
df[['Destination 22 City', 'Destination ID2 City']] = df['Destination ID City'].str.split(', ', expand=True)
df[['Destination 22 Region', 'Destination ID2 Region']] = df['Destination ID Region'].str.split(', ', expand=True)
```
